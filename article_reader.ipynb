```python
def flatten(lists):
    return [x for list in lists for x in list]

def counts(list):
    d = defaultdict(int)
    for x in list: d[x] += 1
    return d

def print_counts(counts):
    for k, c in sorted(counts.iteritems(), key=lambda x: x[1], reverse=True):
        print k, c

def gethost(article):
    return urlparse.urlparse(article['url']).hostname

def sourcename(article):
    return article['site_name'] or gethost(article)
```
